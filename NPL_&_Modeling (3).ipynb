```python
import pandas as pd
df= pd.read_csv("/content/singapore_airlines_reviews.csv")
df.head()
```

```python
df= df.drop(["published_date","type","published_platform","title"], axis=1)
```

## Data Preprocessing

```python
import nltk
```

```python
nltk.download("all")
```

```python
import re
```

```python
stopwords= nltk.corpus.stopwords.words("english")
```

```python
cleaned_records = []
for rec in df.text:
  rec= rec.lower()
  text = re.sub(r'[^a-zA-Z\s]', '', rec)

  words= nltk.tokenize.word_tokenize(rec)
  words= [i for i in words if i not in stopwords]
  text= " ".join(words)

  cleaned_records.append(text)
```

```python
df["cleaned_text"]= cleaned_records
```

```python
df.head()
```

```python
X= df.cleaned_text
y= df['rating']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(X)
```

```python
X_vec= vectorizer.transform(X).toarray()
```

```python
X_vec
```

```python
len(X)
```

```python
X_vec.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_vec, y)
```

## Modeling

```python
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred= model.predict(X_test)
```

```python
pd.DataFrame({"Actual": y_test, "Predictions": y_pred})
```
